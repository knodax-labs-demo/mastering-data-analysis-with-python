```python
import pandas as pd
import sweetviz as sv
from sklearn.datasets import load_iris

# Load the Iris dataset into a DataFrame
iris = load_iris(as_frame=True)
df = iris.frame

# Create an analysis report of the full dataset
report = sv.analyze(df)
report.show_html("sweetviz_iris_report.html")

# Split into train and test datasets
train_df = df.sample(frac=0.8, random_state=42)
test_df = df.drop(train_df.index)

# Compare train and test datasets
compare_report = sv.compare([train_df, "Training"], [test_df, "Testing"])
compare_report.show_html("sweetviz_compare.html")
```
