```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {'Salary': [45000, 48000, 52000, 54000, 56000, 58000, 61000, 65000, 200000]}
df = pd.DataFrame(data)

df['Experience'] = [1, 2, 3, 4, 5, 6, 7, 8, 15]  # Imbalanced pair
plt.scatter(df['Experience'], df['Salary'])
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.title('Experience vs. Salary')
plt.grid(True)
plt.show()


```
