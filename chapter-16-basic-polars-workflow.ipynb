```python
import polars as pl

# Create a DataFrame
df = pl.DataFrame({
    "city": ["London", "Berlin", "Madrid", "Rome"],
    "temperature": [21, 19, 25, 22]
})

# Filter cities with temperature > 20
filtered = df.filter(pl.col("temperature") > 20)

# Show the result
print(filtered)
```
