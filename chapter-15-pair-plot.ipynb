```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset('titanic')
numeric = titanic[['age', 'fare', 'pclass']].dropna()
 
sns.heatmap(numeric.corr(), annot=True, cmap='vlag')
plt.title("Correlation Heatmap - Titanic")
plt.show()
 
sns.pairplot(numeric)
plt.suptitle("Pair Plot - Titanic Data", y=1.02)
plt.show()
```
